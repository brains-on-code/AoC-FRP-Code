## Calculate FRPs

```python
base_frp_parameters = {ERP_PARAMETER_PARTICIPANTS: list(folders.keys()),
                       ERP_PARAMETER_CONDITIONAL_STIMULI: EEG_STIMULUS_SNIPPET_START,
                       ERP_PARAMETER_PLOT: False, ERP_PARAMETER_CORRECT_TRIALS_ONLY: False}
frp_parameter_possibilities = OrderedDict()

frp_parameter_possibilities[ERP_PARAMETER_EPOCH_INTERVAL] = [(-0.3, 1.0)] # ,(-0.3, 1.5),(-0.5, 1.5), (-0.4, 2), (-0.4, 2.5),(-0.6, 3)]
frp_parameter_possibilities[ERP_PARAMETER_ERP_FRP] = [FIXATION_SELECTION_ALGORITHM_FIRST_AOI_PADDING_SMALL]
frp_parameter_possibilities[ERP_PARAMETER_SNIPPET_GROUP] = [SNIPPET_GROUP_ALL]
topomap_times_possibilities = {(-0.6, 3): [0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.8, 2.4, 3.0], (-0.4, 2.5): [
    0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.8, 2.4], (-0.4, 2): [0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8],
    (-0.3, 1.5): [0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1, 1.2, 1.48],
    (-0.5, 1.5): [0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1, 1.2, 1.48],
    (-0.3, 1.0): [0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.98], }
# snippet_groups = {'CalcDiff': [0,10,11,12,13,14,15,30,5,54,7,8,9], 'CalcEqual': [16,17,18,4,49,57,6,61,62,63]}# 60 more for clean, thus left out

all_frp_parameter_combinations = []
for combo in itertools.product(*frp_parameter_possibilities.values()):
    if type(combo) is dict:
        frp_parameters = combo
    else:
        frp_parameters = {list(frp_parameter_possibilities.keys())[
            i]: c for i, c in enumerate(combo)}
    frp_parameters = frp_parameters | base_frp_parameters
    frp_parameters[ERP_PARAMETER_TOPOMAP_TIMES] = topomap_times_possibilities[frp_parameters[ERP_PARAMETER_EPOCH_INTERVAL]]
    frp_parameters[ERP_PARAMETER_CONDITIONAL_STIMULI] = FRP_EEG_STIMULUS_SNIPPET_START[frp_parameters[ERP_PARAMETER_ERP_FRP]]
    all_frp_parameter_combinations.append(frp_parameters)
with open(get_erp_parameters_path(False), 'w') as f:
    json.dump(all_frp_parameter_combinations, f, indent=4)

for frp_parameters in tqdm(all_frp_parameter_combinations):
    print(frp_parameters)
    perform_eeg_erp_averaging(**frp_parameters)
```
