```python
from tqdm.notebook import tqdm
from utils.utils import *
```

```python
# get all files from participants
folders = get_participant_folder_per_participant()
```

```python
statistics_fixations = []
statistics_gazes = []
```

### Check exclusion criteria for visual data 

```python
for participant in tqdm(folders):
    print('----------------------------------------------')
    print(PARTICIPANT, participant)

    print('(01/06) Load exclusions')
    snippet_exclusions: dict[str, list[str]] = get_exclusions(
        participant, [SNIPPET], [VISUAL, EEG, BEHAVIORAL])[SNIPPET]

    print('(02/06) Sort out previously excluded snippets')
    snippet_exclusions = {snippet: snippet_exclusions[snippet][BEHAVIORAL] +
                          snippet_exclusions[snippet][EEG]+snippet_exclusions[snippet][VISUAL] for snippet in snippet_exclusions}
    # remove this step's exclusions to allow multiple successive runs
    snippet_exclusions = {snippet: [exc for exc in snippet_exclusions[snippet] if not exc.startswith('01d8_PRE_VIS_Trial_Exclusion')]
                          for snippet in snippet_exclusions}
    snippets_not_excluded = {snippet: snippet_exclusions[snippet]
                          for snippet in snippet_exclusions if not snippet_exclusions[snippet]}


    print('(03/06) Load fixations, calculate metrics')
    participant_statistics_fixations = calculate_fixation_statistics(participant, snippets_not_excluded)

    print('(04/06) Load gazes, calculate metrics')
    participant_statistics_gazes = calculate_gaze_statistics(participant, snippets_not_excluded)

    print('(05/06) Save statistics data')
    participant_statistics_fixations.to_csv(
        get_statistic_fixations_path(participant), sep=SEPARATOR, index=False)
    statistics_fixations.append(participant_statistics_fixations)
    participant_statistics_gazes.to_csv(
        get_statistic_gazes_path(participant), sep=SEPARATOR, index=False)
    statistics_gazes.append(participant_statistics_gazes)

print('(06/06) Save statistics data for all participants')
pd.concat(statistics_fixations).to_csv(
    get_statistic_fixations_path(), sep=SEPARATOR, index=False)
pd.concat(statistics_gazes).to_csv(
    get_statistic_gazes_path(), sep=SEPARATOR, index=False)
```
